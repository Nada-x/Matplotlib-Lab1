```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
%matplotlib inline
```

## Q1: Create two charts using list and dataframe.

```python
# Creating a bar chart using a list
categories = ['Category A', 'Category B', 'Category C']
values = [10, 20, 15]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()

# Creating a line chart using a DataFrame
data = {'Year': [2010, 2011, 2012, 2013, 2014, 2015],
        'Sales': [100, 120, 90, 150, 200, 180]}

df = pd.DataFrame(data)

plt.plot(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Line Chart')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [30, 20, 15, 35]

plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart', fontsize=20)

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

plt.scatter(x, y, marker='o', alpha=0.4, s=50)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Chart')

plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

plt.scatter(x, y, marker='o', alpha=0.4, s=50)
plt.xlabel('X', fontsize=16)
plt.ylabel('Y', fontsize=16)
plt.title('Scatter Chart', fontsize=18, color='red', loc='left')

plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 4, 3, 2, 1]
y2 = [1, 2, 3, 4, 5]
y3 = [3, 2, 4, 1, 5]

plt.plot(x, y1, label='Line 1', color='blue')
plt.plot(x, y2, label='Line 2', color='purple')
plt.plot(x, y3, label='Line 3', color='green')

# Positioning the legend on the left side
plt.legend(loc='upper left')

plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 4, 3, 2, 1]
y2 = [1, 2, 3, 4, 5]
y3 = [3, 2, 4, 1, 5]

fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(20, 10))

# Plotting the first chart
axes[0].plot(x, y1)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y1')
axes[0].set_title('Chart 1')

# Plotting the second chart
axes[1].plot(x, y2)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y2')
axes[1].set_title('Chart 2')

# Plotting the third chart
axes[2].plot(x, y3)
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y3')
axes[2].set_title('Chart 3')

# Adjusting the layout
plt.tight_layout()

plt.show()
```
